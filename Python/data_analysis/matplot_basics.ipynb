```python
#Sales data visualisation
import pandas as pd

df = pd.read_csv("Sales Data.csv")
df.head()
```

```python
grouped_city = df.groupby('City')['Sales'].agg(["sum", "mean"]).reset_index()
grouped_city
```

```python
plt.figure(figsize=(20,8), facecolor='white')
plt.subplot(1,2,1)
plt.bar(grouped_city['City'], grouped_city["sum"], color='blue', edgecolor='black')
plt.xlabel("City")
plt.ylabel("Total Sales $")
plt.title("Total Sales")
plt.xticks(rotation=90)


# plt.figure(figsize=(20,10), facecolor='white')
plt.subplot(1,2,2)
plt.bar(grouped_city['City'], grouped_city["mean"], color='red', edgecolor='black')
plt.xlabel("City")
plt.ylabel("Average Sales $")
plt.title("Avaerage Sales")
plt.xticks(rotation=90)
plt.ylim(180,190)
```
