```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# basic plotting
tips = sns.load_dataset('tips')
tips
```

```python
## create scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("total bill vs tip")
plt.show()
```

```python
# line plot

sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("line plot of total bill by size")
plt.show()
```

```python
#Categorical plots
#Bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("bar plot of total by day")
plt.show()
```

```python
## Box plot

sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
## distribution of data using violin plot

sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# Histogram
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.show()
```

```python
# KDE(Kernel Density) plot

sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# Pair plot, to check the relationship between total 2 numerical feature.

sns.pairplot(data=tips)
```

```python
# Heat map 
corr=tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
