```python
import pandas as pd
```

```python
csv_path = "Sales Data.csv"
```

```python
df = pd.read_csv(csv_path)
df.head()
```

```python
df.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
df.head()
```

```python
## Handling missing values but there are no missing value.
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
## filling missing value wuith mean of column
df["Price Each"] = df["Price Each"].fillna(df["Price Each"])
df.head(5)
```

```python
df.dtypes
```

```python
# Renaming coloumn
df = df.rename(columns={"Order Date": "Date"})
```

```python
# change datatype for Date to Date time format
df["Date"] = pd.to_datetime(df["Date"]).dt.date
df.head()
```

```python
# Data Aggregating and grouping
sum_sales = df.groupby(['City'])['Sales'].sum().reset_index()
sum_sales = sum_sales.sort_values('Sales', ascending=False)
sum_sales
```

```python
# multiple aggregating function
multiple_grouped = df.groupby('City')["Sales"].agg(["sum", "mean", 'count']).reset_index()
multiple_grouped
```

```python
# Mergin and joining dataframe
df1 = pd.DataFrame({"Key": ["A", "B", "C"], "Value1":[1,2,3]})
df2 = pd.DataFrame({"Key": ["A", "B", "D"], "Value2":[4,5,6]})
df1
```

```python
# Merge 2 df on the key column, inner join
pd.merge(df1,df2,on='Key',how="inner")
```

```python
# Merge 2 df on the key column, outer join
pd.merge(df1,df2,on='Key',how="outer")
```

```python
# Merge 2 df on the key column, left outer join
pd.merge(df1,df2,on='Key',how="left")
```

```python
# Merge 2 df on the key column, right outer join
pd.merge(df1,df2,on='Key',how="right")
```
